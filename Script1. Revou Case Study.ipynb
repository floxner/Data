```python
import pandas as pd
import os
```

```python
# present working directory (meaning lets find inside os get current working directory / folder)
# so its easy to move file around, as long as its in the same folder, so we can move the folder, 
# and the directory still be the same)
pwd = os.getcwd()
```

```python
# have to put \\ on the file path
# dont forget sheet_name (case sensitive)
dataset = pd.read_excel(pwd +"\\Amazon_TRY_GENRE_FINAL.xlsx", sheet_name="Amazon_Data_Edited")
dataset
```

```python
dataset_modified = dataset.copy()
```

```python
# to look of the columns
dataset_modified.columns
```

```python
# make a list so that it can be viewed better
id_vars = list(dataset_modified.columns)[0:16]

value_vars = list(dataset_modified.columns)[16:]
```

```python
# melt or unpivot the data or transpose
# so the first parameter / id_vars is the one that multiply
# the second parameter / value_vars is the one that turn into a row
# the variable is the column, the value is the data / answer
dataset_melted = dataset_modified.melt(id_vars = id_vars, value_vars = value_vars, var_name="Genre", value_name="Genre_List")
dataset_melted
```

```python
# To remove NA
dataset_dropna = dataset_melted.dropna(subset=['Genre_List'])
dataset_dropna
```

```python
#making an excel file to be the output, the name, dont forget \\ , sheet name, and index false, so that the indext doesnt go into the excel
dataset_dropna.to_excel(pwd + "\\Final_Output.xlsx", sheet_name="Final", index=False)
```
